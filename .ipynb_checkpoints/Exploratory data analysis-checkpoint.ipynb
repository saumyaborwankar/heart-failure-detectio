```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset=pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(dataset.corr(), annot = True, cmap='coolwarm',linewidths=.1)
plt.show()
```

# Creatinine Phosphokianse distribution

```python
sns.distplot(dataset['creatinine_phosphokinase'],kde=False,bins=30,color='violet')
```

```python
sns.distplot(dataset['serum_creatinine'],kde=False,bins=30,color='red')
plt.show()
```

# Number of people who have heart disease according to age


```python
plt.figure(figsize=(15,6))
sns.countplot(x='age',data = dataset, hue = 'DEATH_EVENT',palette='GnBu')
plt.show()
```
